```python
sql_query = """
SELECT "Handset Type", COUNT(*) AS handset_count
FROM public.xdr_data
WHERE "Handset Type" IS NOT NULL
 AND "Handset Type" <> 'undefined'
GROUP BY "Handset Type"
ORDER BY handset_count DESC
LIMIT 10;
"""
```

```python

# Execute the SQL query and fetch the results into a DataFrame
top_10_handset_df = pd.read_sql_query(sql_query, conn)
top_10_handset_df
```

```python
sql_query = """
SELECT 
    "MSISDN/Number" AS "User Number",
    COUNT(DISTINCT "Bearer Id") AS "Number of xDR sessions"
FROM public.xdr_data
GROUP BY "MSISDN/Number";
"""
```

```python
# Execute the SQL query and fetch the results into a DataFrame
aggr_xDR_sessions_df = pd.read_sql_query(sql_query, conn)
aggr_xDR_sessions_df
```

```python
sql_query = """
SELECT 
    "MSISDN/Number" AS "User Number",
    COUNT(DISTINCT "Bearer Id") AS "Number of xDR sessions",
    SUM("Dur. (ms)") AS "Total Session Duration (ms)"
FROM public.xdr_data
GROUP BY "MSISDN/Number";
"""
```

```python
# Execute the SQL query and fetch the results into a DataFrame
aggr_session_duration_df = pd.read_sql_query(sql_query, conn)
aggr_session_duration_df
```

```python
sql_query = """
SELECT 
    "MSISDN/Number" AS "User Number",
    COUNT(DISTINCT "Bearer Id") AS "Number of xDR sessions",
    SUM("Total DL (Bytes)") AS "Total Download Data (Bytes)",
    SUM("Total UL (Bytes)") AS "Total Upload Data (Bytes)"
FROM public.xdr_data
GROUP BY "MSISDN/Number";
"""
```

```python
# Execute the SQL query and fetch the results into a DataFrame
total_DL_UL_data_df = pd.read_sql_query(sql_query, conn)
total_DL_UL_data_df
```

```python
sql_query = """
SELECT 
    "MSISDN/Number" AS "User Number",
    "Bearer Id" AS "Session Identifier",
    SUM("Social Media DL (Bytes)" + "Social Media UL (Bytes)") AS "Social Media Total Data Volume (Bytes)",
    SUM("Youtube DL (Bytes)" + "Youtube UL (Bytes)") AS "Youtube Total Data Volume (Bytes)",
    SUM("Netflix DL (Bytes)" + "Netflix UL (Bytes)") AS "Netflix Total Data Volume (Bytes)",
    SUM("Google DL (Bytes)" + "Google UL (Bytes)") AS "Google Total Data Volume (Bytes)",
    SUM("Email DL (Bytes)" + "Email UL (Bytes)") AS "Email Total Data Volume (Bytes)",
    SUM("Gaming DL (Bytes)" + "Gaming UL (Bytes)") AS "Gaming Total Data Volume (Bytes)",
    SUM("Other DL (Bytes)" + "Other UL (Bytes)") AS "Other Total Data Volume (Bytes)",
    SUM("Total DL (Bytes)" + "Total UL (Bytes)") AS "Total Total Data Volume (Bytes)"
FROM public.xdr_data
GROUP BY "MSISDN/Number", "Bearer Id";
"""
```

```python
# Execute the SQL query and fetch the results into a DataFrame
total_data_volume_df = pd.read_sql_query(sql_query, conn)
total_data_volume_df
```
